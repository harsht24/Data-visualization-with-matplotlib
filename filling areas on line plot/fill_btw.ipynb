```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
plt.style.available
```

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
age = data['Age']
all_dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(age, all_dev_salaries, label='All devs', linestyle='--', color='black')
plt.plot(age, py_salaries, label='Python devs')

plt.fill_between(age, py_salaries, alpha=.25)

plt.title('Age vs Developers Salary')
plt.xlabel('Age')
plt.ylabel('Salary')

plt.legend()
plt.tight_layout()
```

```python
plt.plot(age, all_dev_salaries, label='All devs', linestyle='--', color='black')
plt.plot(age, py_salaries, label='Python devs')

median_salary = 57387
plt.fill_between(age, py_salaries, median_salary, alpha=.25)

plt.title('Age vs Developers Salary')
plt.xlabel('Age')
plt.ylabel('Salary')

plt.legend()
plt.tight_layout()
```

```python
plt.plot(age, all_dev_salaries, label='All devs', linestyle='--', color='black')
plt.plot(age, py_salaries, label='Python devs')

median_salary = 57387
plt.fill_between(age, py_salaries, median_salary,
                 where=(py_salaries > median_salary), interpolate=True, alpha=.25)

plt.title('Age vs Developers Salary')
plt.xlabel('Age')
plt.ylabel('Salary')

plt.legend()
plt.tight_layout()
```

```python
plt.plot(age, all_dev_salaries, label='All devs', linestyle='--', color='black')
plt.plot(age, py_salaries, label='Python devs')

median_salary = 57387

plt.fill_between(age, py_salaries, median_salary,
                 where=(py_salaries > median_salary), interpolate=True, alpha=.25, label='Above Average')

plt.fill_between(age, py_salaries, median_salary,
                 where=(py_salaries <= median_salary), interpolate=True, alpha=.25, color='g', label='Below Average')

plt.title('Age vs Developers Salary')
plt.xlabel('Age')
plt.ylabel('Salary')

plt.legend(loc='upper left')
plt.tight_layout()
```
