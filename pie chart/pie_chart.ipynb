```python
import matplotlib.pyplot as plt
```

```python
slices = [59219, 55466, 47544, 36443, 35917] #popularity of programming language.
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.title('Programming language popularity')
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'}) # edgecolor is the color of seperating line.
plt.tight_layout()
```

```python
slices = [59219, 55466, 47544, 36443, 35917] 
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0,0.2]
#javascript is exploded out by 20% from radius.
plt.title('Programming language popularity')

plt.pie(slices, labels=labels, explode=explode, wedgeprops={'edgecolor':'black'})
plt.tight_layout()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0,0.2]
#javascript is exploded out.
plt.title('Programming language popularity')

plt.pie(slices, labels=labels, explode=explode , shadow=True, wedgeprops={'edgecolor':'black'})
# looks better with shadow argument.
plt.tight_layout()
```

```python
slices = [59219, 55466, 47555, 36438, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0,0.2]

plt.title('Programming language popularity')

plt.pie(slices, labels=labels, explode=explode , shadow=True, startangle=60 ,wedgeprops={'edgecolor':'black'})
# startangle rotates the figure by 60 degree.
plt.tight_layout()
```

```python
slices = [59219, 55466, 47555, 36438, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0,0.2]

plt.title('Programming language popularity')

plt.pie(slices, labels=labels, explode=explode , shadow=True, startangle=60 , autopct='%1.1f%%' , wedgeprops={'edgecolor':'black'})
#autopct displays each pie's percentage.
plt.tight_layout()
```
