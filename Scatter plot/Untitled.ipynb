```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y)
plt.tight_layout()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y, s=100, c='b', edgecolor='black', linewidth=1, alpha=0.75 )
plt.tight_layout()
```

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.figure(figsize=(10,6))
plt.scatter(x, y, s=100, edgecolor='black',  c=colors, cmap='Blues')
cbar = plt.colorbar()
cbar.set_label('Satisfaction level')

plt.tight_layout()
```

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.figure(figsize=(10,6))
plt.scatter(x, y, s=sizes, edgecolor='black',  c=colors, cmap='Blues')
cbar = plt.colorbar()
cbar.set_label('Satisfaction level')

plt.tight_layout()
```

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10,6))

plt.scatter(view_count, likes, linewidth=1, alpha=0.75, edgecolor='black')

plt.title('Trending Youtube Video')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(view_count, likes, linewidth=1, alpha=0.75, edgecolor='black')
plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Video')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
```

```python
plt.figure(figsize=(12,8))

plt.scatter(view_count, likes, c=ratio, linewidth=1, alpha=0.75, edgecolor='black')

cbar = plt.colorbar()
cbar.set_label('Like / Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Video')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
```
